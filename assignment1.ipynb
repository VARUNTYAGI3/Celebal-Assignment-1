```python
df["initial_price"] = pd.to_numeric(df["initial_price"], errors="coerce")
df["final_price"] = pd.to_numeric(df["final_price"], errors="coerce")
```

```python
df["discount_amount"] = (df["initial_price"] * df["discount"]) / 100
```
